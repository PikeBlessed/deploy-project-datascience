# Exploratory Analysis of New Dataset 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/home/pikeblessed/proyecto_phnan/deploy-project-datascience/data/dataset_upgraded_phnan.csv', delimiter=';')
df = df.drop(['comments', 'likes', 'saved', 'shares'], axis=1)
df.head(1)
```

```python
df['reach'].describe()
```

```python
df[df['reach'] > 5500].hist(figsize=(10,6), color='brown');
```

```python
sns.histplot(data=df, x='reach', hue='style ', palette='bright', multiple='stack', bins=30)
plt.title('Reach of Content divided for Style')
plt.xlabel('Reach Content')
plt.ylabel('Count posts');
```

```python
sns.histplot(data=df, x='reach', hue='voice_off', palette='bright', multiple='stack')
plt.title('Reach of Content divided for voice_off')
plt.xlabel('Reach Content')
plt.ylabel('Count reels');
```

```python
sns.histplot(data=df, x='reach', hue='main_topic', palette='bright', multiple='stack', bins=30)
plt.title('Reach of Content divided for main_topic')
plt.xlabel('Reach Content')
plt.ylabel('Count posts');
```

```python
sns.scatterplot(data=df, x='duration (seg)', y='reach', hue='voice_off', palette='bright', s=60)
plt.title('Relationship between Reach and Duration')
plt.xlabel('Reel Duration in Segs')
plt.ylabel('Reach Content');
```

```python
sns.pairplot(data=df, hue='main_topic')
```
